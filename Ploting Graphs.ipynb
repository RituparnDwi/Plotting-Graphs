# Bar Graph

```python
# Bar chart or bar graph is a chart or graph that present categorical data with rectangular bars.
# In a Bar graph, while one axis of the chart shows categories being compared, the other axis, on the other hand,
# represent measured values. 
```

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
# syntx- bar(x, height, width, bottom, align)

# x represting x coordinate of the bar
# bottom optional parameter, it denote the Y coordinates of the bar
# align can have value of 'center' or 'edge'. default value is 'center'
```

```python
data = [27, 45, 36, 18, 72, 81, 108, 19, 9]
x = range(1, 10)
plt.bar(x, data)
plt.show()
```

```python
data = [27, 45, 36, 18, 72, 81, 108, 19, 9]
x = range(1, 10)
plt.bar(x, data, align = 'edge')
plt.show()
```

```python
sub = ['Mathematics', 'Computer', 'English', 'Accountancy', 'Economics']
num_of_student = [10, 40, 35, 29, 20]
plt.bar(sub, num_of_student, color = 'green', edgecolor = 'red', linewidth = 2)
# If value of linewidth is zero then no edges will be drawn
plt.show()
```

```python
# Draw horizontel bar graph
```

```python
sub = ['Mathematics', 'Computer', 'English', 'Accountancy', 'Economics']
num_of_student = [10, 40, 35, 29, 20]
plt.barh(sub, num_of_student, color = 'green', edgecolor = 'red', linewidth = 2)
plt.title('No. of Student Interested in Learning Subject') 
plt.show()
```

```python
# plot bar graph, marks obtained by two students in four subjects
```

```python
student1_marks = (83, 79, 80, 83)
student2_marks = (95, 84, 89, 99)
index = np.array([1, 2, 3, 4])
width = 0.30
plt.figure(figsize = (12,6))
plt.bar(index, student1_marks, width, label = 'Marks Obtained by Student1')
plt.bar(index + width, student2_marks, width, label = 'Marks Obtained by Student2')
plt.xticks(index, ['Math', 'Physics', 'Chemistry', 'Hindi'])
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Comparing Marks of Student')
plt.legend(loc='best')
plt.show()
```

### stacked bar graph

```python
# Stacked bar chart stacks bars that represent different groups on top of each other 
```

```python
students = ['Rayeansh', 'Tanvi', 'Kumar', 'Praveen']
Test1 = np.array([18, 14, 12, 15])
Test2 = np.array([20, 18, 16, 19])
Test3 = np.array([16, 16, 13, 15])
index = students
plt.bar(index, Test1, width=0.5, label='Test1', color='red', bottom=Test2 + Test3)
plt.bar(index, Test2, width=0.5, label ='Test2', color='green', bottom=Test3)
plt.bar(index, Test3, width=0.5, label ='Test3', color='blue')
plt.xticks(index, students)
plt.xlabel('Student')
plt.ylabel('Marks')
plt.legend(loc='upper right')
plt.title('Marks Obtained in Test')
plt.show()
```
